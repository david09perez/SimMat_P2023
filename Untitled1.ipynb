```python
#polinomio de grado 'n'
def ajuste(x,y,n):
    #Critero de optimizacion(minimos cuadrados)
    def fun(beta,x,y):
        X=np.array([x**i for i in range(len(beta))])
        f=np.dot(beta,X)
        return np.sum((y-f)**2)
    #solucion optima
    beta0=np.random.rand(n+1)
    soln=minimize(fun,beta0,args=(x,y))
    #grafica
    plt.scatter(x,y,s=5,c='k')
    X=np.array([x**i for i in range(n+1)])
    f=np.dot(soln.x,X)
    plt.plot(x,f,label ='Grado'+str(n))
    plt.legend()
    plt.grid()
    return soln.x,soln.fun
```

```python
s1=ajuste(x,yh,1) #polinomio de grado 1
```

```python
s2=ajuste(x,yl,1) #polinomio de grado 1
```

```python
N=3
soluciones=[]
for i in range (1,N+1):
    soluciones.append(ajuste(x,y,i)) #agregamos al final de la lista el eekemento que regrese 'ajuste'
```
